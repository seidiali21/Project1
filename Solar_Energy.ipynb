```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt1G = pd.read_csv('datas/Plant_1_Generation_Data.csv')
plt2G = pd.read_csv('datas/Plant_2_Generation_Data.csv')
Plt1W = pd.read_csv('datas/Plant_1_Weather_Sensor_Data (1).csv')
Plt2W = pd.read_csv('datas/Plant_2_Weather_Sensor_Data.csv')
```

```python
Plant_Generation_Data = [plt1G, plt2G]
Plant_Generation_Data_result = pd.concat(Plant_Generation_Data)
Plant_Generation_Data_result.head()
```

```python
Plant_Generation_Data_result_clean = Plant_Generation_Data_result

Plant_Generation_Data_result_clean.head()
```

```python
Plant_Generation = pd.DataFrame(Plant_Generation_Data_result_clean, \
                                columns = ['DATE_TIME', 'TOTAL_YIELD']).\
                                set_index('DATE_TIME').sort_values('TOTAL_YIELD',ascending=False).head(20)
Plant_Generation
```

```python
Plant_Generation.plot(kind='barh')
plt.savefig('plot.png')
```

```python
plt.scatter("DATE_TIME", "TOTAL_YIELD")
```

```python
Plant_Weather_Data = [Plt1W, Plt2W]
Plant_Weather_Data_result = pd.concat(Plant_Weather_Data)
Plant_Weather_Data_result.head(20)
```

```python
Plant_Weather_Data_result.head()
```

```python
Plant_Weather_Data_result = Plant_Weather_Data_result.drop(columns = ['IRRADIATION'])
Plant_Weather_Data_result.head()
```

```python
Plant_Weather= pd.DataFrame(Plant_Weather_Data_result, \
                            columns = ['AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE']).\
                            set_index('AMBIENT_TEMPERATURE').sort_values('MODULE_TEMPERATURE', ascending = False).head(20)
Plant_Weather
```

```python
Plant_Weather.plot(kind='barh')
plt.savefig('plot.png')
```

```python
plt.scatter("AMBIENT_TEMPERATURE", "MODULE_TEMPERATURE")
```

```python
plt.xlabel("AMBIENT_TEMPERATURE")
plt.ylabel("MODULE_TEMPERATURE")
plt.show()
```
